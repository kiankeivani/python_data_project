```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
df_US 
```

```python
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_US_pivot
```

```python
df_US_pivot.reset_index(inplace=True)
df_US_pivot
```

```python
df_US_pivot['month_index'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B')
df_US_pivot['month_index']
df_US_pivot.sort_values('month_index', inplace=True)
```

```python
df_US_pivot.set_index('job_posted_month', inplace=True)
df_US_pivot
```

```python
df_US_pivot=df_US_pivot.drop(columns = 'month_index')
df_US_pivot
```

```python
df_US_pivot.plot(kind = 'line')
```

```python
top3 = df_US['job_title_short'].value_counts().head(3)
top3 = top3.index.tolist()
```

```python
df_US_pivot[top3].plot(kind='line')
plt.title("Monthly job postings for top data jobs in the US")
plt.xlabel("2023")
plt.ylabel("job_count")
```

```python
df_merged = df_US_pivot.merge(df_US_software, on='job_posted_month')
```
