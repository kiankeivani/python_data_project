```python
import pandas as pd
from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df_job_country_salary = df.pivot_table(values = 'salary_year_avg', index = 'job_country',columns = 'job_title_short', aggfunc= 'median')
```

```python
top_countries = df['job_country'].value_counts().head(6).index
top_countries
```

```python
targeted_jobs = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_job_country_salary = df_job_country_salary.loc[top_countries, targeted_jobs]
```

```python
import matplotlib.pyplot as plt
df_job_country_salary.plot(kind = 'bar')
plt.ylabel("Median salary ($USD)")
plt.xlabel("")
plt.title("Median Salary by Top 6 Countries")
plt.xticks(rotation=45, ha = 'right')
plt.show()
```
