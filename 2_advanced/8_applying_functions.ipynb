```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df_salary = df[df['salary_year_avg'].notna()].copy()
df_salary2 = df[df['salary_year_avg'].notna()].copy()
```

```python
def inflated_salary(salary):
    return salary * 1.03
df_salary['inf_salary'] = df_salary['salary_year_avg'].apply(inflated_salary)
df_salary[['inf_salary', 'salary_year_avg']]
```

```python
df_salary2['inf_salary'] = df_salary2['salary_year_avg'].apply(lambda salary: salary * 1.03)
df_salary2[['inf_salary', 'salary_year_avg']]
```
